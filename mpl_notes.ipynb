# MatPlotLib
- Makes it easy to plot points on a graph
- Typically used for all 2D representation

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-100,100,50)    # plot 50 points from -100 to 100 x-axis
y = x ** 2
plt.plot(x, y, 'g--', linewidth=3)
plt.xlabel ('$x$', fontsize=18)
plt.ylabel ('$y^2$', fontsize=18)

plt.show()
```

```python
x = np.linspace(-1.5, 1.3, 50)
plt.plot(x, x, 'b--', x, x**2, 'g:', x, x ** 3, 'm^', linewidth=2)
```
